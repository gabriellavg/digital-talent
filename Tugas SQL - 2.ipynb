# Tugas SQL - 2

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

### One to many relationship

__1. Report the account representative for each customer.__

```python
run ("""
SELECT 
    customerName AS CustomerName,
    CONCAT(firstName, ' ', lastName) AS SalesRepName
FROM
    customers
        INNER JOIN
    employees
        ON
    salesRepEmployeeNumber = employeeNumber
ORDER BY
    customerName
""")
```

__2. Report total payments for Atelier graphique.__

```python
run ("""
SELECT 
    SUM(amount) AS TotalPayments
FROM
    payments
        INNER JOIN
    customers
        USING
    (customerNumber)
WHERE
    customerName = 'Atelier graphique'
""")
```

__3. Report the total payments by date__

```python
run ("""
SELECT
    paymentDate AS PaymentDate,
    SUM(amount) AS TotalPayments
FROM
    payments
GROUP BY
    PaymentDate
""")
```

__4. Report the products that have not been sold.__

```python
run ("""
SELECT
    productCode AS ProductCode,
    productName AS ProductName
FROM
    products
        LEFT JOIN
    orderdetails
        USING
    (productCode)
WHERE
    orderdetails.productCode IS NULL
""")
```

__5. List the amount paid by each customer.__

```python
run ("""
SELECT
    customerName AS CustomerName,
    SUM(amount) AS AmountPaid
FROM
    customers
        INNER JOIN
    payments
        USING
    (customerNumber)
GROUP BY
    customerNumber
""")
```

__6. How many orders have been placed by Herkku Gifts?__

```python
run ("""
SELECT
    COUNT(orders.customerNumber) AS TotalOrders
FROM
    orders
        INNER JOIN
    customers
        USING
    (customerNumber)
WHERE customerName = 'Herkku Gifts'
""")
```

__7. Who are the employees in Boston?__

```python
run ("""
SELECT
    CONCAT(firstName, ' ', lastName) AS EmployeeName
FROM
    employees
        INNER JOIN
    offices
        USING
    (officeCode)
WHERE 
    city = 'Boston'
""")
```

__8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.__

```python
run ("""
SELECT 
    customerName AS CustomerName,
    SUM(amount) AS TotalPayments
FROM 
    customers
        INNER JOIN
    payments
        USING 
    (customerNumber)
GROUP BY
    customerNumber
HAVING
    TotalPayments > 100000
ORDER BY
    TotalPayments DESC
""")
```

__9. List the value of 'On Hold' orders.__

```python
run ("""
SELECT
    orderNumber AS OrderNumber,
    productCode AS ProductCode,
    quantityOrdered AS QuantityOrdered,
    priceEach AS PriceEach,
    QuantityOrdered * PriceEach AS Value
FROM
    orders
        INNER JOIN
    orderdetails
        USING
    (orderNumber)
WHERE
    status = 'On Hold'
""")
```

__10. Report the number of orders 'On Hold' for each customer.__

```python
run ("""
SELECT
    customerName AS CustomerName,
    COUNT(orders.customerNumber) AS TotalOnHoldOrders
FROM
    customers
        INNER JOIN
    orders
        USING
    (customerNumber)
WHERE
    status = 'On Hold'
GROUP BY
    orders.customerNumber
""")
```

### Many to many relationship

__1. List products sold by order date.__

```python
run ("""
SELECT
    orderDate AS OrderDate, 
    productName AS ProductName
FROM
    orderdetails
        INNER JOIN
    orders
        USING
    (orderNumber)
        INNER JOIN
    products
        USING
    (productCode)
ORDER BY
    OrderDate
""")
```

__2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.__

```python
run ("""
SELECT
    orderDate AS OrderDate
FROM
    orderdetails
        INNER JOIN
    orders
        USING
    (orderNumber)
        INNER JOIN
    products
        USING
    (productCode)
WHERE
    productName = '1940 Ford Pickup Truck'
ORDER BY
    OrderDate DESC
""")
```

__3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?__

```python
run ("""
SELECT
    customerName AS CustomerName,
    orderNumber AS OrderNumber
FROM
    orderdetails
        INNER JOIN
    orders
        USING
    (orderNumber)
        INNER JOIN
    customers
        USING
    (customerNumber)
GROUP BY 
    OrderNumber
HAVING
    SUM(quantityOrdered * priceEach) > 25000
""")
```

```python
run ("""
SELECT
    productName AS ProductName,
    COUNT(orderNumber) AS TotalOrders
FROM
    orders
        INNER JOIN
    orderdetails
        USING
    (orderNumber)
        RIGHT JOIN
    products
        USING
    (productCode)
GROUP BY 
    products.productCode
ORDER BY
    TotalOrders DESC
""")
```

__5. List the names of products sold at less than 80% of the MSRP.__

```python
run ("""
SELECT DISTINCT
    productName AS ProductName,
    MSRP,
    priceEach AS PriceEach
FROM
    orderdetails
        INNER JOIN
    products
        USING
    (productCode)
WHERE 
    PriceEach < 0.8 * MSRP
""")
```

__6. Reports those products that have been sold with a markup of 100% or more (i.e.,  the priceEach is at least twice the buyPrice)__

```python
run ("""
SELECT DISTINCT
    productName AS ProductName,
    buyPrice AS BuyPrice,
    priceEach AS PriceEach
FROM
    orderdetails
        INNER JOIN
    products
        USING
    (productCode)
WHERE 
    PriceEach >= 2 * BuyPrice
""")
```

__7. List the products ordered on a Monday.__

```python
run ("""
SELECT DISTINCT
    productName AS ProductName
FROM
    orderdetails
        INNER JOIN
    orders
        USING
    (orderNumber)
        INNER JOIN
    products
        USING
    (productCode)
WHERE
    DAYNAME(orderDate) = 'Monday'
""")
```

__8. What is the quantity on hand for products listed on 'On Hold' orders?__

```python
run ("""
SELECT
    products.productName AS ProductName,
    quantityInStock AS QuantityInStock
FROM
    orderdetails
        INNER JOIN
    orders
        USING
    (orderNumber)
        INNER JOIN
    products
        USING
    (productCode)
WHERE
    status = 'On Hold'
""")
```
